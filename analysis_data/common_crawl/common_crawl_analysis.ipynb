```python
import json
import os
import pandas as pd
from tqdm import tqdm

# 讀取 JSONL 檔案並轉換成 DataFrame
def jsonl_to_dataframe(file_path):
    data = []
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            try:
                # 解析 JSON 字串並加入到列表中
                data.append(json.loads(line))
            except json.JSONDecodeError:
                print("Skipping invalid JSON line:", line)
                continue
    # 將列表轉換成 DataFrame
    df = pd.DataFrame(data)
    return df

# 主程式
if __name__ == "__main__":
    parent_dir = '/home/u3659277/master-thesis/analysis_data/common/'
    temp = []
    for path in tqdm(os.listdir(parent_dir)):
        temp.append(jsonl_to_dataframe(os.path.join(parent_dir,path))) 
    
    df_concat = pd.concat(temp, ignore_index=True)
    # 將 JSONL 檔案轉換成 DataFrame
    # df1 = jsonl_to_dataframe(jsonl_file_path)

    # 輸出 DataFrame 的前幾列
    # print(df1.head())

```

```python
def process_sentence(sentence, target_pattern, attribute_pattern, remove_pattern):
    if re.findall(remove_pattern, sentence): return None
    target_matches = re.findall(target_pattern, sentence)
    if not target_matches: return None
    attribute_matches = re.findall(attribute_pattern, sentence)
    if not attribute_matches: return None
      
    combinations = [(t,a) for t in set(target_matches) for a in set(attribute_matches) if a not in t]
        
    return combinations if combinations else None

def collect_biased_sentences_6(sentences_data, target_terms, attribute_terms, remove_target_terms, sentence_ids):
    # 对目标术语和属性术语使用集合去重
    target_terms_set = set(target_terms)
    attribute_terms_set = set(attribute_terms)
    target_pattern = '|'.join(target_terms_set)
    remove_pattern = '|'.join(remove_target_terms)
    attribute_pattern = '|'.join(attribute_terms_set)
    # 使用 apply 处理每个句子
    t_a_combinations = sentences_data.swifter.apply(lambda s: process_sentence(s, target_pattern, attribute_pattern, remove_pattern))

    # 过滤掉没有组合的句子
    filtered_results = t_a_combinations[t_a_combinations.notnull()]

    df_biased_sentences = pd.DataFrame({
        'Sentence ID': sentence_ids[filtered_results.index],
        'Biased Sentences': sentences_data[filtered_results.index],
        'T-A Combination': filtered_results
    })

    return df_biased_sentences
```

```python
from rank_bm25 import BM25Okapi

# 定義函式，使用BM25計算相似度並找出相似度超過閥值的句子
def find_similar_sentences(df, df_compare, threshold):
    # 將第一群句子轉換為列表
    df_sentences_list = df.values.tolist()
    
    # 將第二群句子轉換為列表
    df_compare_sentences_list = df_compare.values.tolist()
    
    # 創建BM25模型
    bm25 = BM25Okapi(df_compare_sentences_list)
    
    # 查看分数范围
    scores = bm25.get_scores(df_sentences_list)
    min_score = min(scores)
    max_score = max(scores)
    
    print(min_score,max_score)
    
    # 儲存相似句子的列表
    similar_sentences = []
    
    # 對於第二群句子中的每個句子，計算其與第一群句子的相似程度
    for sentence in df_compare_sentences_list:
        # 計算相似度
        scores = bm25.get_scores(sentence)
        # 檢查相似度是否超過閥值
        if max(scores) > threshold:
            similar_sentences.append(sentence)
    
    return similar_sentences

# 設定閥值
threshold = 200  # 請根據實際情況調整閥值

# 使用BM25算法找出相似句子
similar_sentences = find_similar_sentences(compare_df['Biased Sentences'],sampled_data['sentence'], threshold)



```
